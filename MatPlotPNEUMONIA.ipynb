```python
import cv2
import glob
from glob import glob
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from random import shuffle
#image = mpimg.imread("chelsea-the-cat.png")
```

```python
#Set Library Directory then shuffle
path = glob(r"/Users/jeffreyscruggs/Desktop/DataSets/chest_xray/shuffle/*.*")
shuffle(path)
```

```python
#Cycle through directory while predicting!
import time
for file in path:
    print(file)
    a = mpimg.imread(file)
    test_image = image.load_img(file, target_size = (150,150))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)
    result = model.predict(test_image)
    print(result)
    
    if result == 0:
        print("This is a NORMAL scan.")
        #img = Image.open(file)
        
    
        
        plt.imshow(a)
        plt.text(400, 400, 'NORMAL', bbox=dict(fill=False, edgecolor='blue', linewidth=2))
        plt.show()
        
        #Wait a 2 seconds
        time.sleep(2)
        plt.close()
    else:
        print("This scan has PNEUMONIA.")
        #img = Image.open(file)
        
        plt.imshow(a)
        plt.text(400, 400, 'PNEUMONIA', bbox=dict(fill=False, edgecolor='red', linewidth=2))
        plt.show()
        time.sleep(2)
        plt.close()
        #Destroy Windows to save memory.
```
